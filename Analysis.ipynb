```python
def make_table1(results):
    # Make a table with all values from:
        # DPWGAN with ϵ = 1
        # WGAN
    results_2=[]
    lables_1=["DPWGAN_01.00","WGAN"]
    for x in lables_1:
        for y in lables[0]:
            results_2+=[values(lists_with(results,[x,y]))]
    multi_index = pd.MultiIndex.from_product([["AUROC"],["DPWGAN ϵ = 1","WGAN"], lables[0]])
    index = pd.MultiIndex.from_product([["Without Boosting","With Boosting"],lables[2]], names=['', 'Model'])
    df = pd.DataFrame(transpose(results_2), index=index, columns=multi_index)
    display(df)
    
make_table1(results)
```

```python
# Plotting function
def plot_fig_by_ind(ind=2,title=None,results=None, ax=None, ind2=1):
    # Plots AUROC vs ϵ split on a feature
    names = lables[ind]
    x_axis = lables[ind2]
    if ind  == 3: names=["Without boosting", "With boosting"]
    if ind2 == 1: x_axis  = [str(x) for x in [0.01, 0.05, 0.1, 0.5, 1, 5, 10, "inf"]]
    for i,x in enumerate(lables[ind]):
        vals=[ave(values(lists_with(results,[x,y]))) for y in lables[ind2]]
        if sum(vals)==0: continue
        ax.plot(x_axis,vals,label=names[i])
        ax.set_xlabel("ϵ")
        ax.set_ylabel("AUROC")
    ax.legend(loc='best', framealpha=0.5)
    if not title: title = "Average Accuracy Over All Models on " + results[0][0] + "." #
    ax.set_title(title)
```

```python
# Create figures for report
def plot_fig_1_2_3():
    fig, ax = plt.subplots(3, 1, figsize=(12, 16))
    plot_fig_by_ind(3,"Figure 1. Average Accuracy Over All Models With and Without Boosting.",results=results,ax=ax[0])
    plot_fig_by_ind(2,"Figure 2. Average Accuracy Over All Models by Model Type.",results=results,ax=ax[1])
    plot_fig_by_ind(0,"Figure 3. Average Accuracy Over All Models by Dataset ",results=results,ax=ax[2])
    plt.tight_layout()
    plt.show()
plot_fig_1_2_3()
```
